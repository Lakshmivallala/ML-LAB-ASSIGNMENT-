## Problem 1

```python
import numpy as np
import math
from sklearn import datasets
from sklearn.datasets import load_iris
from matplotlib import pyplot as plot
```

```python
x=np.array([5,3,4])
y=np.array([1,8,6])
n=x.size
a=0
```

### Euclidean Distance

```python
for i in range(n):
    a = a + (x[i] - y[i])**2
b=(a)**(1/2)
print("Euclidean distance is",b)
```

### Manhattan Distance

```python
a=0
for i in range(n):
    a = a + abs(x[i]-y[i])
print("Manhattan distance is:",a)
```

### Minkowski Distance

```python
a=0
q=3
for i in range(n):
    a = a + (abs(x[i]-y[i]))**q
b=a**(1/q)
print("Minkowski distance is:",b)
```

### Euclidean distance between two rows

```python
def ED(row1,row2):
    a=0.0
    for i in range(4):
        a = a + (row1[i]-row2[i])**2
    return math.sqrt(a)
```

```python
for i in range(np.shape(data.data)[0]):
    for j in range(np.shape(data.data)[0]):
         matrix[i][j]=ED(data.data[i],data.data[j])
    print(matrix) 
```
